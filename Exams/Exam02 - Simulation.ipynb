```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
%matplotlib inline
```

```python
# simuler 100 tirages d'une loi normale ¬µ=3 et ùúé = 0.1
simu_norm= np.random.normal(loc=3, scale=0.1, size=1000)

import matplotlib.pyplot as plt
# (c) Histogramme
plt.hist(simu_norm, bins=30, edgecolor="black")
plt.title("Histogramme de simu_norm")
plt.xlabel("Valeurs")
plt.ylabel("Fr√©quence")
plt.show()
```

```python
# Ins√©rez votre code ici
moyenne_simu = simu_norm.mean()
ecart_type_simu = simu_norm.std()

print(f"Moyenne : {moyenne_simu}")
print(f"Ecart Type : {ecart_type_simu}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Param√®tres th√©oriques
mu_theorique = 3
sigma_theorique = 0.1

# Diff√©rentes tailles d'√©chantillon
tailles = np.array([10, 50, 100, 500, 1000, 5000, 10000, 100000, 1000000, 10000000])

# Simulation
np.random.seed(42)
moyennes = []
ecarts_types = []

print("Convergence vers les valeurs th√©oriques :")
print("="*60)
for taille in tailles:
    echantillon = np.random.normal(loc=mu_theorique, scale=sigma_theorique, size=taille)
    moy = echantillon.mean()
    std = echantillon.std()

    moyennes.append(moy)
    ecarts_types.append(std)

    print(f"n={taille:5d} ‚Üí moyenne={moy:.6f}, √©cart-type={std:.6f}")

print("="*60)
print(f"Valeurs th√©oriques : Œº={mu_theorique}, œÉ={sigma_theorique}")

# Visualisation
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.semilogx(tailles, moyennes, 'bo-', markersize=8)
plt.axhline(y=mu_theorique, color='red', linestyle='--', linewidth=2, label='Valeur th√©orique')
plt.xlabel('Taille √©chantillon (√©chelle log)')
plt.ylabel('Moyenne')
plt.title('Convergence de la moyenne')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.semilogx(tailles, ecarts_types, 'go-', markersize=8)
plt.axhline(y=sigma_theorique, color='red', linestyle='--', linewidth=2, label='Valeur th√©orique')
plt.xlabel('Taille √©chantillon (√©chelle log)')
plt.ylabel('√âcart-type')
plt.title('Convergence de l\'√©cart-type')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
