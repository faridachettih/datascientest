```python
#(h)  Transformez-la en variable numérique en supprimant la sous-chaîne de caractères "-month"
def clean_duration(x):
    return int(x[:-6])

df["Duration"] = df["Duration"].apply(clean_duration)
# On converti le type des valeurs en int
df["Duration"] = pd.to_numeric(df["Duration"])
df.head()
```

```python
#(i) création duration à 3 modalités
def Duration_categ(d):
    if d <= 10:
        return "court-terme"
    elif d <= 30:
        return "moyen-terme"
    else:
        return "long-terme"

df["Duration_categ"] = df["Duration"].apply(Duration_categ)
df.head()
```

```python
#(j)
# 1) Filtrer les locataires
df[df['Housing'] == 'rent']['Duration_categ'].value_counts().index[0]
```

```python
#(k)
saving_account_mapping = {"little": 0, "moderate": 1, "quite rich": 2, "rich": 3}
df["Saving accounts"] = df["Saving accounts"].replace(saving_account_mapping)
df.head()
```

```python
#(l)
checking_account_mapping = {"little": 0, "moderate": 1, "quite rich": 2, "rich": 3}
df["Checking account"] = df["Checking account"].replace(checking_account_mapping)
df.head()
```

```python
#(m) les doublons
df.duplicated().any()
```

```python
#(n)
df.isna().sum()
```

```python
df['Housing'].value_counts()
```

```python
#(o) trouver le mode de la colonne
mode_housing = df["Housing"].mode()[0]
#remplacer les valeurs manquantes
df["Housing"] = df["Housing"].fillna(mode_housing)
```

```python
#(p)
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
#(q)
pred_german = pd.read_csv('predictions_german.csv', index_col=0)
```

```python
#(r)
df_pred = pd.concat([df, pred_german], axis=1)
```

```python
#(s) création de la cellule 'error' en utilisant la soustraction
df_pred['error'] = df_pred['duration'] - df_pred['predictions']
```

```python
#(t)
df_pred[df_pred['error'] < 0]['error'].count()
```
